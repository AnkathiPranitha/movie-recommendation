```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('/content/drive/MyDrive/IITBHU/netflix_titles - Copy.csv')
```

```python
def get_important_features(data):
  important_features = []
  for i in range(0, data.shape[0]):
    important_features.append(data['cast'][i]+' '+data['director'][i]+' '+data['description'][i]+' '+data['title'][i])

  return important_features
```

```python
df['important_features'] = df['cast'].astype(str) + ' ' + df['director'].astype(str) + ' '+ df['description'].astype(str) + ' '+df['title']
df
```

```python
cm = CountVectorizer().fit_transform(df['important_features'])
```

```python
cs = cosine_similarity(cm)
print(cs)
```

```python
cs.shape
```

```python
Title = 'The Conjuring'
```

```python
Show_id = df[df.title == Title]['show_id'].values[0]
```

```python
scores = list(enumerate(cs[Show_id]))
```

```python
sorted_scores = sorted(scores, key = lambda x:x[1], reverse = True)
sorted_scores = sorted_scores[1:]
```

```python
print(sorted_scores)
```

```python
j=0
print('The 7 most recommended movies to',Title, 'are:\n')
for item in sorted_scores:
 movie_title = df[df.show_id == item[0]]['title'].values[0]
 print(j+1, movie_title)
 j = j+1
 if j>6:
    break
```
